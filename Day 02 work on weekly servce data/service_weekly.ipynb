## service wise summary admit vs refused

```python
service_summary = services_weekly.groupby('service')[['patients_admitted','patients_refused']].sum().reset_index()
service_summary['Total_Requests'] = service_summary['patients_admitted'] + service_summary['patients_refused']
service_summary['Admit Percentage'] = (service_summary['patients_admitted'] / service_summary['Total_Requests']) * 100

fig1 = px.pie(
    service_summary,
    values='Total_Requests',
    names='service',
    color_discrete_sequence=px.colors.qualitative.Set3,
    title='Overall Service Distribution (Total Patients Request)',
    hole=0.3
)
fig1.update_traces(textposition='inside', textinfo='percent+label')
fig1.show()
```

## weekly admit vs refused

```python
weekly_summary = services_weekly.groupby('week')[['patients_admitted', 'patients_refused']].sum().reset_index()
fig2 = px.bar(
    weekly_summary,
    x='week',
    y=['patients_admitted','patients_refused'],
    barmode='group',
    title='weekly admitted vs refused',
    text_auto = True    
)
fig2.show()
```

## highest satisfaction ratio by service

```python

satisfaction = services_weekly.groupby('service')['patient_satisfaction'].mean().reset_index()
fig3 = px.bar(
    satisfaction,
    x='patient_satisfaction',
    y='service',
    orientation='h',
    color='patient_satisfaction',
    color_continuous_scale='Blues',
    title='Average Patient Satisfaction by Service',
    
)
fig3.show()
```

## bed Utilization rate (admitted/available)

```python
services_weekly['bed Utilization'] = (services_weekly['patients_admitted'] / services_weekly['available_beds']) * 100
bed_Util = services_weekly.groupby('service')['bed Utilization'].mean().reset_index()
fig4 = px.bar(
    bed_Util,
    x='service',
    y='bed Utilization',
    color='bed Utilization',
    color_continuous_scale='Viridis',
    title='Average Bed Utilization Rate Per Service',
    text_auto='.1f'
       
    
)
fig4.show()
```

## Staff moral trends by weeks

```python
morale_trend = services_weekly.groupby('week')['staff_morale'].mean().reset_index()
fig5=px.line(
    morale_trend,
    x='week',
    y='staff_morale',
    markers=True,
    title='Staff Morale Trend over a Time',
    line_shape='spline'
    
)
fig5.show()
```
